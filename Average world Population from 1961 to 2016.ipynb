```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv('https://tinyurl.com/ChrisCoDV/world_population.csv', index_col=0)


data
```

```
data= data.transpose()
print(data.head())
```

We dont need some of the rows. So, we are dropping the rows. 

```
#Drop rows

selected= ['Country Code', 'Indicator Name', 'Indicator Code']

data= data.drop(selected, axis=0)
print(data)



```

```
#Sorting 

data=data.reindex(data.mean().sort_values(ascending=False).index, axis=1)
```

```
print(data.mean())
```

**China , Monaco and Hongkong are the largest three country in therms of average population density with 15403, 14643 and 5370 people respectively.**

```
select2= data.columns[data.mean()> 5000]

data= data[select2]

print(data)


data.plot.line()
plt.show()
```
